```python
import numpy as np
import pandas as pd
from IPython.display import clear_output
try:
    row_num = 6
    col_num = 7
    def space_generate():
        space_field = np.zeros((row_num,col_num))
        return space_field

    def drop_start(space_field, row, col, entry_data):
        space_field[row][col] = entry_data


    def validation(space_field, col):
        return space_field[row_num-1][col] == 0


    def free_row(space_field, col):
        for r in range(row_num):
            if space_field[r][col] == 0:
                return r

    def check_win(space_field, entry_data):
        for c in range(col_num-3):
            for r in range(row_num):
                if space_field[r][c] == entry_data and space_field[r][c+1] == entry_data and space_field[r][c+2] == entry_data and space_field[r][c+3] == entry_data:
                    return True

        for c in range(col_num):
            for r in range(row_num-3):
                if space_field[r][c] == entry_data and space_field[r+1][c] == entry_data and space_field[r+2][c] == entry_data and space_field[r+3][c] == entry_data:
                    return True

        for c in range(col_num-3):
            for r in range(row_num-3):
                if space_field[r][c] == entry_data and space_field[r+1][c+1] == entry_data and space_field[r+2][c+2] == entry_data and space_field[r+3][c+3] == entry_data:
                    return True

        for c in range(col_num-3):
            for r in range(3, row_num):
                if space_field[r][c] == entry_data and space_field[r-1][c+1] == entry_data and space_field[r-2][c+2] == entry_data and space_field[r-3][c+3] == entry_data:
                    return True



    def print_field(space_field):
        li_game_op = (np.flip(space_field,0)).tolist()
        output = [str(cc).replace("1.0","Y").replace("0.0","_").replace("2.0","R").replace(",","") for cc in li_game_op]
        for row in output:
            print(*row)
        if counter_op_clear >0:
                clear_output(wait=True)




    counter_op_clear = 0
    print("Start to play New GAME")
    space_field = space_generate()
    print_field(space_field)
    game_over = False
    turn = 0
    flag = 2
    while not game_over:
        flag = 2
        counter_op_clear = 1
        if turn == 0:
            col = int(input("Player Yellow Make Selection between 0-6\n"))
            if (col>-1 and col<=6) and validation(space_field,col):
                row = free_row(space_field,col)
                drop_start(space_field,row,col,1)
                if check_win(space_field,1):
                    #print("\nYellow Wins\n")
                    print_field(space_field)
                    print("\nYellow Wins\n")
                    ch = input("Enter 1 for Restart\nEnter any other key for Exit\n")
                    clear_output(wait=True)
                    if ch == '1':
                        counter_op_clear = 0
                        #clear_output(wait=True)
                        print("Let's play New GAME")
                        space_field = space_generate()
                        game_over = False
                        turn = 1
                        flag = 2
                    else:
                        game_over = True
                        clear_output(wait=True)
                        print("Game Exit\nThank You")
                        break
            else:
                print("You have entered In-valid choice Please enter valid choice")
                flag = 0
        else:
            col = int(input("Player Red Make Selection between 0-6\n"))
            if (col>-1 and col<=6) and validation(space_field,col):
                row = free_row(space_field,col)
                drop_start(space_field,row,col,2)
                if check_win(space_field,2):
                    #print("\nRed Wins\n")
                    print_field(space_field)
                    print("\nRed Wins\n")
                    ch = input("Enter 1 for Restart\nEnter any other key for Exit\n")
                    clear_output(wait=True)
                    if ch == '1':
                        counter_op_clear = 0
                        #clear_output(wait=True)
                        print("Let's play New GAME")
                        space_field = space_generate()
                        game_over = False
                        turn = 1
                        flag = 2
                    else:
                        game_over = True
                        clear_output(wait=True)
                        print("Game Exit\nThank You")
                        break
            else:
                print("You have entered In-valid choice Please enter valid choice")
                flag = 1          

        print_field(space_field)
        if flag == 0:
            turn = 0
            #flag = 2
        elif flag == 1:
            turn = 1
            #flag = 2
        else:
            turn += 1
            turn %= 2
            #flag = 2

except:
    clear_output(wait=True)
    print("Game Exit\nThank You")
```
